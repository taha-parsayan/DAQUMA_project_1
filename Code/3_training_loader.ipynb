# Make loader files from the excel files
They need to have the technical names as the header and save as csv

```python
import os
import pandas as pd
import numpy as np

current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)

loader_path = os.path.join(parent_dir, 'Data', 'Clean_data', 'Biotheus', 'Training_Nantong', 'Loader')
```

Load the excel files

```python
training_data_path = os.path.join(parent_dir, 'Data', 'Clean_data', 'Biotheus', 'Training_Nantong', 'Excel', 'Training.xlsx')
training_df = pd.read_excel(training_data_path, sheet_name=None) # Load all sheets

LR = training_df['Learner Role']
Cur = training_df['Curriculum']
TR = training_df['Training Requirement']
Link_C_LR = training_df['Link C-LR']
Link_C_TR = training_df['Link C-TR']
```

Learner Role

```python
# Rename columns
LR.rename(columns={
    'Name': 'name__v',
    'Unit': 'site__c',
    'Scope': 'scope__c',
    'RegArea': 'reg_area__c',
    'Description': 'description__v'
}, inplace=True)

# Create external_id__c by prefixing 'MIG-' to name__v
LR['external_id__c'] = 'MIG-' + LR['name__v']

# Save to CSV
LR.to_csv(os.path.join(loader_path, 'loader_Learner_Role.csv'), index=False)
```

Curriculum

```python
# Rename columns
Cur.rename(columns={
    'Name': 'name__v',
    'Unit': 'site__c',
    'Owning Department':'department__c',
    'RegArea': 'reg_area__c',
    'Scope': 'scope__c',
    'Description': 'description__v'
}, inplace=True)

# Create external_id__c by prefixing 'MIG-' to name__v
Cur['external_id__c'] = 'MIG-' + Cur['name__v']

# Save to CSV
Cur.to_csv(os.path.join(loader_path, 'loader_Curriculum.csv'), index=False)
```

Training Requirement

```python
# Rename columns
TR.rename(columns={
    'Name': 'name__v',
    'Unit': 'site__c',
    'Training Requirement Type':'object_type__v',
    'Required Quiz': 'require_quiz__v',
    'Duration': 'duration__v',
    'Duration Unit': 'duration_unit__v',
    'Estimated Time to Complete (Minutes)': 'estimated_time_minutes__v',
    'Recur': 'is_recurring__v',
    'Recur Frequency (months)': 'recurrence__v',
    'Description': 'description__v'
}, inplace=True)

TR['recurrence_unit__v'] = 'Months'

# Create external_id__c by prefixing 'MIG-' to name__v
TR['external_id__c'] = 'MIG-' + TR['name__v']

# Save to CSV
TR.to_csv(os.path.join(loader_path, 'loader_Training_Requirement.csv'), index=False)
```

Link C-LR

```python
# Rename columns
Link_C_LR.rename(columns={
    'Curriculum': 'curriculum__v.external_id__c',
    'Learner Role Name': 'learner_role__v.external_id__c'
}, inplace=True)

# Prefix 'MIG-' to external IDs
Link_C_LR['curriculum__v.external_id__c'] = 'MIG-' + Link_C_LR['curriculum__v.external_id__c']
Link_C_LR['learner_role__v.external_id__c'] = 'MIG-' + Link_C_LR['learner_role__v.external_id__c']

# Save to CSV
Link_C_LR.to_csv(os.path.join(loader_path, 'loader_Link_C_LR.csv'), index=False)
```

Link_C_TR

```python
# Rename columns
Link_C_TR.rename(columns={
    'Curriculum': 'curriculum__v.external_id__c',
    'Training Requirement': 'training_requirement__v.external_id__v'
}, inplace=True)

# Prefix 'MIG-' to external IDs
Link_C_TR['curriculum__v.external_id__c'] = 'MIG-' + Link_C_TR['curriculum__v.external_id__c']
Link_C_TR['training_requirement__v.external_id__v'] = 'MIG-' + Link_C_TR['training_requirement__v.external_id__v']

# Save to CSV
Link_C_TR.to_csv(os.path.join(loader_path, 'loader_Link_C_TR.csv'), index=False)
```
